# Scatter Plot

## Overview

This section introduces the scatter plot. A scatter plot can be very useful in order to visiualize how the data behaves or how two variables may be correlated. 

## Scatter Plot

There is not much to say about a scatter plot. So let's see how to generate such a plot with Python.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x, alpha, beta, normal_dist):
    return alpha * x + beta + normal_dist
```

```python
np.random.seed(42)

mu = 2.0
std = 1.0

x = np.linspace(0.0, 6.0, 1000)
sample = np.random.normal(mu, std, size=1000)

y = func(x=x, alpha = 2.0, beta = 1.0, normal_dist=sample)

plt.scatter(x, y, label='scatter')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple scatter plot')
plt.show()
```

## Summary

In this section we saw how to generate a scatter plot with Python.
