# Simulation Methods


Simulation methods play a key role in modern computational statistics. In this section, we discuss some 
common methods frequently used in statistics. Namely, Monte Carlo integration, Metropolis-Hastings algorithm 
and importance sampling. We begin however, with simple coin flips and the Binomial distribution.
Then we introduce the <a href="https://en.wikipedia.org/wiki/Bootstrapping_(statistics)">boostrap algorithm</a>.
