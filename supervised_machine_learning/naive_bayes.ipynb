A classical classificatin problem is  to classify emails into spam or not spam. Another example may be to classify amazon reviews into positive or negative sentiment.
One way to do so is to use a  <a href="https://en.wikipedia.org/wiki/Naive_Bayes_classifier">naive Bayes</a> classifier.  This is perhaps one of the most straightforward classification algorithms and it is based on <a href="https://en.wikipedia.org/wiki/Bayes'_theorem">Bayes theorem</a>.

## Naive Bayes

Naive Bayes has many applications like sentiment analysis, author identification, information retrieval and word disambiguation. 
Overall, it is a simple, fast and robust algorithm. Given these properties, naive Bayes is frequently used as a baseline model. 

Consider  two events $A$ and $B$ the naive Bayes theorem is experssed as follows

$$P(A | B) = P(B | A)\frac{P(A)}{P(B)}$$

The naive Bayes method has certain advantages, some of which have already been mentioned. 

The following are the core disadvantages of the naive Bayes approach. The assumption of independent features is not always true. Indeed, it is almost impossible that we  will have a set of predictors that are fully independent. In addition, if no training data exists for a particular class, this is transalted as a zero posterior probability. In this scenario,
the model cannot make any predictions. This is also known as the zero probability problem.


### Laplacian smoothing

<a href="https://en.wikipedia.org/wiki/Additive_smoothing">Laplacian smoothing</a> (or additive smoothing) is a technique we can use in order to avoid the probability going to zero. This is the case when a data point does not appear in the training set.

### Log likelihood

The likelihood function for the training set is, under the naive Bayes assumprion, the product of probabilities of the data points in the data set. However, this calculation may result in a numerical underflow. We can avoid this by taking the logarightm of the likelihood function.

## Numerical example

Let's see an example of applying naive Bayes using scikit. 
This example is taken from <a href="https://www.datacamp.com/tutorial/naive-bayes-scikit-learn">Naive Bayes Classification Tutorial using Scikit-learn</a>. In the example below, we use ```GaussianNB```. This class implements the Gaussian Naive Bayes algorithm for classification. In this case the likelihood of the features is assumed to be Gaussian that is:

\begin{equation}
p(x_i | y) = \frac{1}{(2 \pi \sigma^{2}_{y})^{1/2}}exp\left(- \frac{1}{2\sigma^{2}_{y}} (x_i - \mu_y)^2 \right)
\label{eq:eq1}\tag{1}
\end{equation}

````python
from sklearn.datasets import make_classification

X, y = make_classification(
    n_features=6,
    n_classes=3,
    n_samples=800,
    n_informative=2,
    random_state=1,
    n_clusters_per_class=1,
)
````

````python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=y, marker="*")
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=125
)
````

````python
from sklearn.naive_bayes import GaussianNB

# Build a Gaussian Classifier
model = GaussianNB()

# Model training
model.fit(X_train, y_train)

# Predict Output
predicted = model.predict([X_test[6]])

print("Actual Value:", y_test[6])
print("Predicted Value:", predicted[0])
````
